# K-means

K-means es un método de agrupamiento, que tiene como objetivo la partición de un conjunto de n observaciones en k grupos en el que cada observación pertenece al grupo cuyo valor medio es más cercano. 

```python
from sklearn.cluster import KMeans
import random
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
X=[]
Y=[]

for i in range (3):
    circulo_r = random.random() * 100

    circulo_x = random.random() * 100
    circulo_y = random.random() * 100



    for i in range(500):

        alpha = 2 * math.pi * random.random()
        r = circulo_r * math.sqrt(random.random())
        x = r * math.cos(alpha) + circulo_x
        y = r * math.sin(alpha) + circulo_y

        X.append(x)
        Y.append(y)
        
d = np.column_stack((X, Y))

plt.scatter(X,Y)
plt.show()


```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(d)
Z = kmeans.predict(d)
centroides = kmeans.cluster_centers_
etiquetas = kmeans.labels_


plt.scatter(X,Y)

plt.plot(centroides[:,0],centroides[:,1],'mo',markersize=8, label='centroides')
plt.legend(loc='best')
plt.show()
```
